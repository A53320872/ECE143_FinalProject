```python
from historical_peaks import get_spot_distribution
import matplotlib.pyplot as plt

dist_2000 = get_spot_distribution(2000)
dist_2019 = get_spot_distribution(2019)

plt.subplot(1, 2, 1)
plt.pie(dist_2000, autopct='%1.0f%%', colors=['rgb(220, 20, 60)', 'rgb(60, 179, 113)', '#FFC133', 'rgb(220, 220, 220)'])
plt.title('Year 2000')

plt.subplot(1, 2, 2)
plt.pie(dist_2019, autopct='%1.0f%%',  colors=['rgb(220, 20, 60)', 'rgb(60, 179, 113)', '#FFC133', 'rgb(220, 220, 220)'])
plt.title('Year 2019')
plt.legend(['Admin (A)', 'Grad (B)', 'Student (S)', 'Visitor (V)'], loc = 'lower left', bbox_to_anchor=(-0.4, -.24, 0.5, 0.5))

plt.show()
```
